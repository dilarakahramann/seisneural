```python
import pandas as pd
from sklearn.preprocessing import MinMaxScaler
from sklearn.model_selection import train_test_split
# 1. Önceden temizleyip kaydettiğin yeni CSV dosyasını okuyoruz
df = pd.read_csv('yeni_ysadata.csv')
# 2. Veri Standardizasyonu (Min-Max Scaling)
# Hedef değişkenimiz (Magnitude) hariç girdi sütunlarını 0-1 arasına çekiyoruz
scaler = MinMaxScaler()
girdi_sutunlari = ['Year', 'Month', 'Day', 'Latitude', 'Longitude', 'Depth']
# Okuduğumuz df tablosunun üzerindeki girdi sütunlarını normalize ediyoruz
df[girdi_sutunlari] = scaler.fit_transform(df[girdi_sutunlari])
# 3. Eğitim ve Test Setlerinin Ayrılması (%80 Eğitim, %20 Test)
X = df[girdi_sutunlari]
y = df['Magnitude']

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.20, random_state=42)
# 4. Sonuçları ekrana yazdırıp kontrol ediyoruz
print("--- NORMALİZE EDİLMİŞ TEMİZ TABLO ÖRNEĞİ ---")
print(df.head())
print("\n--- VERİ SETİ BÖLÜNME DETAYLARI ---")
print(f"Eğitim setindeki girdi (X_train) satır sayısı: {X_train.shape[0]}")
print(f"Test setindeki girdi (X_test) satır sayısı: {X_test.shape[0]}")
```

```python
import joblib
# 1. Normalize edilmiş bütün tabloyu yeni bir CSV olarak kaydedelim
df.to_csv('yeni_ysadata_normalize.csv', index=False)
# 2. Arkadaşlarının modelleri eğitebilmesi için bölünen setleri ayrı ayrı CSV olarak kaydedelim
X_train.to_csv('X_train.csv', index=False)
X_test.to_csv('X_test.csv', index=False)
y_train.to_csv('y_train.csv', index=False)
y_test.to_csv('y_test.csv', index=False)
# 3. Kullandığımız 'scaler' (MinMax) modelini kaydedelim (Arayüz kodlanırken Enes'e lazım olacak)
joblib.dump(scaler, 'deprem_scaler.pkl')

print("--- TÜM VERİLER BAŞARIYLA KAYDEDİLDİ ---")
print("Klasörünüzde şu dosyalar oluştu:")
print("- yeni_ysadata_normalize.csv (Bütün temiz veri)")
print("- X_train.csv ve y_train.csv (Eğitim paketleri)")
print("- X_test.csv ve y_test.csv (Test paketleri)")
print("- deprem_scaler.pkl (Ölçekleyici modeli)")
```
